Pour connaître la médiane du nombre de pages des demandes d'accès colligées dans mon fichier.

```python
acces.nbPages.median()
```

Je veux avoir une idée rapide des 10 ministères/instances gouvernementales qui ont été les plus sollicités par des demandes d'accès à l'information.

```python
acces.nom_ministère.value_counts().head(10)
```

Pour connaître le pourcentage des demandes d'accès accaparés par les 10 ministères/instances gouvernementales les plus sollicités.

```python
acces.nom_ministère.value_counts().head(10) / len(acces.nom_ministère)*100
```

Combien de pages de demandes d'accès cumulent les 10 ministères/instances gouvernementales les plus sollicités?

```python
acces.groupby("nom_ministère").nbPages.sum().sort_values(ascending=False).head(10)
```

J'attribue à la variable "zero" toutes les demandes d'accès qui ont été refusées parce qu'il n'y avait pas de documentation à ce sujet au ministère. J'utilie la partie en anglais de la colonne "réponse" parce que la partie en français est parfois évacuée.

```python
zero = acces.réponse == "No records exist"
```

Sur les 28 588 demandes d'accès colligées dans mon fichier, 173 d'entre elles ont été refusées parce qu'il n'y avait "aucun document existant".

```python
zero.value_counts()
```

Je veux connaître les 10 ministères qui ont le plus refusé de demandes d'accès parce qu'ils n'avaient pas de documentation à ce sujet. Arrive en tête de liste Services publics et Approvisionnement Canada.

```python
acces[zero].nom_ministère.value_counts().sort_values(ascending=False).head(10)
```
