Avec la manipulation ci-dessous, je peux avoir une idée rapide des ministères qui ont été les plus sollicités par des demandes d'accès à l'information.

```python
acces.nom_ministère.value_counts()
```

Je connais le nombre de demandes d'accès adressées auprès de chaque ministère du gouvernement canadien, mais non le pourcentage de chacun.

```python
acces.nom_ministère.value_counts() / len(acces.nom_ministère)*100
```

Je fais afficher les 10 ministères les plus sollicités par des demandes d'accès, avec le nombre dedites demandes pour chacun. Contrairement à l'exercice que nous avons fait en classe le 22 mars, il ne semble pas avoir de répétition dans le nom des ministères (et donc pas de nettoyage nécessaire avec Google Refine)

```python
acces.nom_ministère.value_counts().head(10)
```

Combien de pages de demandes d'accès cumulent les ministères canadiens sollicités?

```python
acces.groupby("nom_ministère").pages.sum().sort_values(ascending=False).head(10)
```

J'attribue à la variable "zero" toutes les demandes d'accès qui ont été refusées parce qu'il n'y avait pas de documentation à ce sujet au ministère. J'utilie la partie en anglais de la colonne «disposition" parce que la partie en français est parfois évacuée.

```python
zero = acces.disposition == "No records exist"
```

```python
acces[zero]
```

```python
zero.value_counts()
```

Je veux connaître les 10 ministères qui ont le plus refusé de demandes d'accès parce qu'ils n'avaient pas de documentation à ce sujet. Je remarque que le ministère des Services publics et Approvisionnement Canada est le grand champion de l'opacité dans ce cas-ci.

```python
acces[zero].nom_ministère.value_counts().sort_values(ascending=False).head(10)
```

Je crée une variable qui regroupe toutes les demandes d'accès qui ont simplement été refusées sans en préciser les raisons ("All excluded").

```python
refus = acces.disposition == "All excluded"
```

Je fais afficher les données de ma nouvelle variable, et je remarque que le Public Services and Procurement Canada est surreprésenté. Encore une fois.

```python
acces[refus]
```

Quel est le champion de la transparence de la demande d'accès? Créons une variable pour le savoir!

```python
champion = acces.disposition == "All disclosed"
```

```python
champion.value_counts()
```

Emploi et Développement social Canada est le ministère du gouvernement canadien qui a le plus répondu par l'affirmative aux demandes d'accès qui lui ont été adressées.
